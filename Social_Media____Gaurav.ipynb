# **3.15 Location-Based Sentiment Heatmap**

```python
# Create a pivot table for location sentiment
location_sentiment_heatmap = df.pivot_table(index='location', columns='sentiment', aggfunc='size', fill_value=0)

# Normalize the data for better visualization
location_sentiment_normalized = location_sentiment_heatmap.div(location_sentiment_heatmap.sum(axis=1), axis=0)

# Plot heatmap for location-based sentiment
sns.heatmap(location_sentiment_normalized, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title(('Location-Based Sentiment Heatmap'),fontsize=20)
plt.xlabel(('Sentiment'),fontsize=15)
plt.ylabel(('Location'),fontsize=15)
plt.show()
```

# **4. Advanced Keyword Analysis**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Function to display top words for each sentiment
def display_top_words(tfidf_matrix, feature_names, sentiment):
    print(f"\nTop words for {sentiment.capitalize()} Sentiment:")
    for i in range(len(feature_names)):
        print(f"{feature_names[i]}: {tfidf_matrix[i]:.4f}")

# Prepare TF-IDF for each sentiment category
for sentiment in ['positive', 'negative', 'neutral']:
    sentiment_texts = df[df['sentiment'] == sentiment]['text']

    tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=10)
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentiment_texts)

    # Summarize the TF-IDF scores
    tfidf_sum = tfidf_matrix.sum(axis=0).tolist()[0]
    feature_names = tfidf_vectorizer.get_feature_names_out()

    # Display top words
    display_top_words(tfidf_sum, feature_names, sentiment)
```

## **5. Convert sentiment to numeric values for correlation analysis**

```python
sentiment_map = {'positive': 1, 'neutral': 0, 'negative': -1}
df['sentiment_numeric'] = df['sentiment'].map(sentiment_map)
# Calculate correlation matrix
correlation_matrix = df[['sentiment_numeric', 'likes', 'retweets']].corr()
# Display correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```
